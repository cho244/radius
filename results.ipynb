```python
import pandas as pd
import csv
%matplotlib inline
import datetime
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from pandas.tools.plotting import scatter_matrix
import numpy as np
```

```python
df = pd.read_csv('test.tsv', sep='\t', header = None)
```

```python
df[2].unique()
```

```python
#properly name column
df['user_id']=df[0]
df = df.drop(0, axis=1)

#properly name column and change date data type
df['activity_date'] = pd.to_datetime(df[1], format = '%Y-%m-%d')
df = df.drop(1, axis=1)

#properly name column
df['activity_type'] = df[2]
df = df.drop(2, axis=1)


#organize dataframe to boolean statements
df['email_opened']=(df['activity_type']=='EmailOpen')
df['form_submit']=(df['activity_type']=='FormSubmit')
df['email_clicked']=(df['activity_type']=='EmailClickthrough')
df['customer_support']=(df['activity_type']=='CustomerSupport')
df['page_view']=(df['activity_type']=='PageView')
df['web_visit']=(df['activity_type']=='WebVisit')

#make into bool statement
df['email_opened']=df.email_opened.astype('int')
df['form_submit']=df.form_submit.astype('int')
df['email_clicked']=df.email_clicked.astype('int')
df['customer_support']=df.customer_support.astype('int')
df['page_view']=df.page_view.astype('int')
df['web_visit']=df.web_visit.astype('int')

gk = df.groupby(['user_id'])
ratios = gk.agg(np.sum)
```

```python
target = ratios.sort_values(by='email_opened' , ascending=False)
target.head(1000)
```

```python
people = []
for person in target:
    if len(people) == 1000:
        break
    else:
        people.append(person)
people
```
